```python
#DFT (Fourier)
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
X = np.fft.fft(x)
x_rec = np.fft.ifft(X)

print("DFT:", X)
print("IDFT:", x_rec.real)
```

```python
#DCT (Cosine)
from scipy.fft import dct, idct

x = np.array([1, 2, 3, 4])
X = dct(x, type=2, norm='ortho')
x_rec = idct(X, type=2, norm='ortho')

print("DCT:", X)
print("IDCT:", x_rec)
```
